```python
import matplotlib.pyplot as plt

# Data
x_labels = ['CC200', 'CC400', 'Harvard-Oxford', 'Dosenbach-160']
y_values = [69.176, 66.806, 58.689, 59.848]

# Light pastel bar colors
bar_colors = ['#FFD699', '#CC99FF', '#FFB6C1', '#D2B48C']

# Create a bar chart with adjusted width and gap
bars = plt.bar(x_labels, y_values, width=0.3, align='center')  # Adjust width as needed

# Assign different colors to each bar
for bar, color in zip(bars, bar_colors):
    bar.set_color(color)

# Adding labels and title with different colors
plt.xlabel('Atlases', color='#3399FF')  # Bright blue color for x-axis label
plt.ylabel('Accuracy', color='brown')  # Vibrant green color for y-axis label
plt.title('L-SVM', color='red')

# Set a specific y-axis range with a 2.5 difference
plt.ylim(min(y_values) - 1.5, max(y_values) + 1.5)

# Add a legend
plt.legend(bars, x_labels, title='Atlases', loc='upper right')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['CC200', 'CC400', 'Harvard-Oxford', 'Dosenbach-160']
accuracy = [0.8668, 0.8533, 0.8312, 0.8473]
acc_std = [2.38, 2.35, 2.73, 1.57]
sensitivity = [0.8683, 0.8633, 0.9404, 0.9406]
f1_score = [0.8579, 0.8453, 0.8379, 0.8510]
auc_score = [0.9571, 0.9531, 0.9509, 0.9515]

# Set lighter colors for each label
accuracy_color = '#a6cee3'  # light blue
sensitivity_color = '#b2df8a'  # light green
f1_score_color = '#fb9a99'  # light red
auc_score_color = '#cab2d6'  # light purple
acc_std_color = 'gray'

# Bar graph for Accuracy, Sensitivity, F1-Score, AUC Score
bar_width = 0.2
index = np.arange(len(labels))

fig, ax1 = plt.subplots(figsize=(8, 5))

rects1 = ax1.bar(index - bar_width, accuracy, bar_width, label='Accuracy', color=accuracy_color, edgecolor='black')
rects2 = ax1.bar(index, sensitivity, bar_width, label='Sensitivity', color=sensitivity_color, edgecolor='black', alpha=0.7)
rects3 = ax1.bar(index + bar_width, f1_score, bar_width, label='F1-Score', color=f1_score_color, edgecolor='black', alpha=0.5)
rects4 = ax1.bar(index + 2 * bar_width, auc_score, bar_width, label='AUC Score', color=auc_score_color, edgecolor='black', alpha=0.3)

# Line graph for ACC.Std
ax2 = ax1.twinx()
ax2.plot(index, acc_std, marker='o', linestyle='-', color='blue', label='ACC.Std', linewidth=2)


# Set labels and title
#ax1.set_xlabel('Atlas')
#ax1.set_ylabel('Metrics Scores')
#ax2.set_ylabel('ACC.Std', color=acc_std_color)
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(labels)
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5)  # Place legend below x-axis in horizontal layout
ax2.legend(loc='upper left', bbox_to_anchor=(0, 0.9))
ax1.set_title('Performance Comparison among Atlases using DNN')

# Format y-axis values to show two decimal points
ax1.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))

# Set y-axis range
ax1.set_ylim([0.78, 0.97])

# Display the plot
plt.show()
```
