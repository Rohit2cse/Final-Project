```python
import matplotlib.pyplot as plt
import numpy as np

# Data
labels = ['CC200', 'CC400', 'Harvard-Oxford', 'Eickhoff-Zilles atlas']
accuracy = [0.8768, 0.8433, 0.8212, 0.8373]
acc_std = [2.38, 2.35, 2.73, 1.57]
sensitivity = [0.8683, 0.8633, 0.9404, 0.9406]
f1_score = [0.8579, 0.8453, 0.8379, 0.8510]
auc_score = [0.9571, 0.9531, 0.9509, 0.9515]

# Set colors for each metric
accuracy_color = '#1f78b4'  # dark blue
sensitivity_color = '#33a02c'  # dark green
f1_score_color = '#e31a1c'  # dark red
auc_score_color = '#6a3d9a'  # dark purple
acc_std_color = 'blue'  # blue

# Bar graph for Accuracy, Sensitivity, F1-Score, AUC Score
bar_width = 0.2
index = np.arange(len(labels))

fig, ax1 = plt.subplots(figsize=(8, 5))

rects1 = ax1.bar(index - bar_width, accuracy, bar_width, label='Accuracy', color=accuracy_color, edgecolor='black')
rects2 = ax1.bar(index, sensitivity, bar_width, label='Sensitivity', color=sensitivity_color, edgecolor='black', alpha=0.7)
rects3 = ax1.bar(index + bar_width, f1_score, bar_width, label='F1-Score', color=f1_score_color, edgecolor='black', alpha=0.5)
rects4 = ax1.bar(index + 2 * bar_width, auc_score, bar_width, label='AUC Score', color=auc_score_color, edgecolor='black', alpha=0.3)

# Line graph for ACC.Std
ax2 = ax1.twinx()
ax2.plot(index, acc_std, marker='o', linestyle='-', color=acc_std_color, label='ACC.Std', linewidth=2)

# Set labels and title
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(labels)
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5)  # Place legend below x-axis in horizontal layout
ax2.legend(loc='upper left', bbox_to_anchor=(0, 0.9))
ax1.set_title('Performance Comparison among Atlases using DNN')

# Format y-axis values to show two decimal points
ax1.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2f'))

# Set y-axis range
ax1.set_ylim([0.78, 0.97])

# Display the plot
plt.show()
```
