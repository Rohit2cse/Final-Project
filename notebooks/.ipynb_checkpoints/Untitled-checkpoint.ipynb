```python
from nilearn import datasets

# Fetch the Harvard-Oxford cortical and subcortical atlas with atlas_version=2
parcellations = datasets.fetch_atlas_harvard_oxford(atlas_name='cort-maxprob-thr50-1mm', data_dir=None, symmetric_split=False, resume=True, verbose=1)

# Get the filename of the atlas
atlas_filename = parcellations['maps']
```

```python
from nilearn import plotting as nlp
import matplotlib.pyplot as plt
nlp.plot_roi(atlas_filename, draw_cross=True)
plt.gcf().set_size_inches(10, 5)
```
